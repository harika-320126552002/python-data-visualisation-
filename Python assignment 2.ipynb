```python
import pandas as pd
import matplotlib.pyplot as plt
iris = pd.read_csv("Iris.csv") # the iris dataset is now a Pandas DataFrame

#To view Iris data below:
iris.head()
```

Description:
In the cell we have read the CSV file using pandas and we printed its head.

```python
plt.rcParams["figure.figsize"] = (8,5)#to increase size of plot
plt.scatter(iris['SepalLengthCm'],iris['SepalWidthCm'],color="black",marker="1",s=80)#s is used for marker size
plt.title("Scatter plot for Sepal and Petal lengths")
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
#plt.grid()
plt.show()
```

Description:

In the above cell we have plotted the scatter plot for SepalLength Vs SepalWidth. We have used the marker as + and black color for the marker.

From this plot we can say that the density is more at 5.5 cm to 7.0 cm

```python
plt.rcParams["figure.figsize"] = (10,6)#to increase size of plot
plt.scatter(iris['SepalLengthCm'],iris['SepalWidthCm'],color="black",marker="1",label="Sepal",s=70)
plt.scatter(iris['PetalLengthCm'],iris['PetalWidthCm'],color="green",marker="*",label="Petal",s=60)
plt.title("Scatter plot for Sepal and Petal lengths")
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.legend()
plt.grid()
```

Description:

In the above cell we have plotted the scatter plot for SepalLength Vs SepalWidth along with PetalLength vs Petalwidth. We have used the marker 1,* and color black, green for the above plots respectively .

From this plot we can say that the density of both petals and sepals almost same amount of ratio of sepallength and sepalwidth but the ratio of petallength and petalwidth is somewhat disturbutive.

```python
import statistics as stat
x=list(iris['Species'])
y=list(iris['SepalLengthCm'])
k=stat.mean(y[0:50])
l=stat.mean(y[51:101])
n=stat.mean(y[101:])
z=[x[0],x[51],x[101]]
m=[k,l,n]
plt.pie(m,labels=z,radius=1.8,label="")
plt.title("Species VS SepalLengthCm")
plt.show()
```

Description:

In the above cell we have plotted the pie chart for average SepalLengths of various Species.

Through pie chart we can intrepret that there no significant difference between the sepellengths of the species.

```python
y=list(iris['PetalLengthCm'])
k=stat.mean(y[0:50])
l=stat.mean(y[51:101])
n=stat.mean(y[101:])
z=[x[0],x[51],x[101]]
explode=[0.1,0,0.05]
m=[k,l,n]
plt.pie(m,labels=z,radius=1,explode=explode)
plt.title("Species VS PetalLengthCm")
plt.show()
```

Description:

In the above cell we have plotted the pie chart for average PetalLengths of various Species.

Through pie chart we can intrepret that Iris-virginica has maximum average PetalLength and Iris-setosa has very less average PetalLength.

```python
y=list(iris['PetalWidthCm'])
k=stat.mean(y[0:50])
l=stat.mean(y[51:101])
n=stat.mean(y[101:])
z=[x[0],x[51],x[101]]
explode=[0.1,0,0.05]
m=[k,l,n]
plt.pie(m,labels=z,radius=1.8,explode=explode)
plt.title("Species VS PetalWidthCm")
plt.show()
```
